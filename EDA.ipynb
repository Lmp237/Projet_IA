```python
def pourcentage_valeurs_manquantes(df):
    return df.isnull().mean() * 100

pourcentage = pourcentage_valeurs_manquantes(df)
resultat = pourcentage.sort_values(ascending=False)
resultat
```

```python
supp = resultat[resultat >= 65].index
clean = df.drop(columns=supp)
clean
```

# Détection des outliers

```python
col = 'AMT_CREDIT'
Q1 = clean[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1
sb = Q1 - 1.5 * IQR
sh = Q3 + 1.5 * IQR

ok = clean[(clean[col] >= sb) & (clean[col] <= sh)]
ok
```

# Etude de la distribution de la target

```python
# Comptage des classes 0 (pas de défaut) et 1 (défaut)
counts = clean['TARGET'].value_counts()
print(counts)

# Visualisation par diagramme en barres
counts.plot(kind='bar')
plt.title('Distribution de la variable TARGET')
plt.xlabel('Classes (0 = non défaut, 1 = défaut)')
plt.ylabel('Nombre d\'échantillons')
plt.show()
```

```python
# Moyenne de 'AMT_INCOME_TOTAL' selon TARGET
moyennes = clean.groupby('TARGET')['AMT_CREDIT'].mean()
print(moyennes)

# Visualisation Boxplot pour comparer les distributions de revenus
clean.boxplot(column='AMT_CREDIT', by='TARGET')
plt.title('Revenu total selon TARGET')
plt.suptitle('')
plt.xlabel('TARGET')
plt.ylabel('AMT_CREDIT')
plt.show()
```
